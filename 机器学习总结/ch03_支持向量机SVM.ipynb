# 支持向量机SVM

支持向量机就是将向量从低维空间映射到高维空间中，在空间中建立一个`最大间隔超平面`，分开的超平面两侧分别建有两个相互平行的临界超平面，临界平面间隔越大，误差越小。

所谓的`支持向量`就是在临界平面上的点。

## 算法原理

给定训练样本集：$D = \{(\pmb{x_1}, y_1), (\pmb{x_2}, y_2), \cdots, (\pmb{x_m}, y_m)\}, y_i∈\{-1, +1\}$，在样本空间中，划分超平面可以使用如下线性方程来描述：

$$
    \pmb{w}^T\pmb{x} + b = 0
$$

其中b是平面关于原点的偏离，W则是超平面的法向量。

任意点到平面的距离计算就是

$$
    distance(\pmb{x},b,\pmb{w}) = |\frac{\pmb{w}^T}{\|\pmb{w}\|}(\pmb{x}-\pmb{x^{'}})| = \frac{1}{\|\pmb{w}\|}|\pmb{w}^T\pmb{x}+b|
$$

由于支持向量在临界超平面上，那么$\boldsymbol{w} ^ T\boldsymbol{x} + b = \pm 1$，也就是说支持向量到平面的距离为：
$$
    γ = \frac{2}{\|\pmb{w}\|}
$$

那么我们求平面间隔最大值，就转化为求$\|\pmb{w}\|$的最小值，等价于
$$
min_{w, b} \frac{1}{2} {\|\pmb{w}\|}^2
$$

$$
s.t. \quad y_i(\pmb{w} ^ T\pmb{x_i} + b) \geq 1, i = 1, 2, \cdots, m
$$

由于目标函数是凸函数，主要的求解方法是使用拉格朗日乘子法求解，具体推导不赘述。

## SVM优势

适用于小样本、非线性及高维数据模式识别等特征较多的应用场合。
* 算法针对有限样本进行训练，目的是获得现有信息下的最优解，而不是无穷样本下的最优解
* 最终将其转化为的凸优化问题，理论上可以寻找到全局最优解。
* 虽然算法是将数据从低维空间映射到高维空间中，但是算法的复杂性与`数据维度无关`。

## 核函数

以上是针对训练样本线性可分的情况下进行的分析，但现实中很少存在线性可分的数据，于是需要将数据从低维映射到高维特征空间中。

但是高维空间中由于高维难以直接计算最佳分类平面，所以使用低维空间的核函数来代替高维特征空间的内积运算，即将x变化为$φ(\pmb{x})$就可以。

> 需要注意的是：核函数和映射是不一样的，核函数是将数据映射到高维，再将数据投影到原始维度，本质上并没有增加数据的维度。

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, svm

%matplotlib inline
# %matplotlib widget
plt.rcParams["font.sans-serif"] = 'SimHei'  # 中文问题
plt.rcParams["axes.unicode_minus"] = False  # 负号问题
%config InlineBackend.figure_format = 'svg'
```

```python
# we create 40 separable points
X, y = datasets.make_blobs(n_samples=40, centers=2, random_state=6, cluster_std=0.80)

# fit the model, don't regularize for illustration purposes
clf = svm.SVC(kernel='linear', C=1000)
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Wistia)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```

```python
def make_meshgrid(x, y, h=0.02):
    """创建要绘制的网格点
    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional
    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    """绘制分类器的决策边界
    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out


iris = datasets.load_iris()
# Take the first two features. We could avoid this by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0  # SVM regularization parameter
models = (
    svm.SVC(kernel="linear", C=C),
    svm.LinearSVC(C=C, max_iter=10000),
    svm.SVC(kernel="rbf", gamma=0.7, C=C),
    svm.SVC(kernel="poly", degree=3, gamma="auto", C=C),
)
models = (clf.fit(X, y) for clf in models)

# title for the plots
titles = (
    "SVC with linear kernel",
    "LinearSVC (linear kernel)",
    "SVC with RBF kernel",
    "SVC with polynomial (degree 3) kernel",
)

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm,
               s=20, edgecolors="k", alpha=0.5)
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel("Sepal length")
    ax.set_ylabel("Sepal width")
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```
