# DBSCAN聚类算法

DBCSAN（Density-Based Spatial Clustering of Applications with Noise）是基于密度的聚类算法，根据密度来计算样本之间的相似度，因此可以挖掘任意形状的簇。

并且能够有效过滤噪声样本对聚类结果的影响，所以可以用作异常检测。

在[DBSCAN可视化](https://www.naftaliharris.com/blog/visualizing-dbscan-clustering/)中可以观察到DBSCAN一步一步的的聚类过程。

## DBSCAN基本原理

DBSCAN是基于中心的密度建立的，样本的密度通过核心对象在半径r内的样本点个数（密度阈值）来估计。那么我们需要设置合适的`半径r`和`密度阈值MinPts`来训练模型。

![image.png](images/DBSCAN_01.png)

* 核心对象：样本x在半径为r的邻域内具有大于MinPts个样本对象，则x是核心对象。
* 直接密度可达：y在以核心对象x为半径r的邻域内，则y与x直接密度可达。
* 密度可达：y在以核心对象x为半径r的邻域内，且z在以核心对象y为半径r的邻域内，则x与z密度可达。
* 密度相连：对于对象x和y，若存在z使x和y是从z关于(r，MinPts)密度可达，那么x和y密度相连。

DBSCAN算法随机选择一个核心对象作为“种子”，根据密度可达关系求出所有密度相连样本的最大集合，将其视为一个簇，反复执行该过程，直到所有的核心对象都属于一个簇。剩下的对象则是异常点。

## DBSCAN 特点

* 优势：
    * 可以发现任意形状的簇
    * 擅长找到异常点
* 劣势
    * 对数据集每个对象进行邻域检查，数据集较大时，聚类收敛时间长（先做降维工作）
    * 参数影响很大
    * 空间聚类的密度不均匀，聚类的质量较差

K-means是基于划分的聚类算法，该算法只能挖掘`球状簇`，对于不规则形状的簇则使用DBSCAN比较合适，DBSCAN也在各个实验中与其他聚类算法相比具有较大优势占据较大优势。

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

%matplotlib inline
# %matplotlib widget
plt.rcParams["font.sans-serif"] = 'SimHei'  # 中文问题
plt.rcParams["axes.unicode_minus"] = False  # 负号问题
%config InlineBackend.figure_format = 'svg'
```

```python
# 按照中心点生成数据，并标准化
centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=0)
X = StandardScaler().fit_transform(X)

# 训练模型
db = DBSCAN(eps=0.3, min_samples=10).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# 簇和噪声的数量
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print('估计聚类数：%d' % n_clusters_)
print('估计噪声点数：%d' % n_noise_)
# 外部指标
print("同质性Homogeneity: %0.3f"
      % metrics.homogeneity_score(labels_true, labels))  # 正确率
print("完整性Completeness: %0.3f"
      % metrics.completeness_score(labels_true, labels))  # 每个簇正确分类的样本占类型的总样本数
print("V-measure: %0.3f"
      % metrics.v_measure_score(labels_true, labels))  # 完整性和同质性的调和平均
print("调整兰德系数Adjusted Rand Index: %0.3f"
      % metrics.adjusted_rand_score(labels_true, labels))  # 描述随机分配簇类标记向量的相似度问题
print("调整的互信息指数Adjusted Mutual Information: %0.3f"
      % metrics.adjusted_mutual_info_score(labels_true, labels))  # 基于预测簇向量与真实簇向量的互信息分数来衡量其相似度的
# 内部质量评价指标
print("轮廓系数Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(X, labels))  # 轮廓系数处于[-1,1]的范围内，-1表示错误的聚类，1表示高密度的聚类，0附近表示重叠的聚类；

# 绘制图像
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # 黑色噪声点
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=14, alpha=0.6)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6, alpha=0.6)

plt.title('估计聚类数: %d' % n_clusters_)
plt.show()
```
