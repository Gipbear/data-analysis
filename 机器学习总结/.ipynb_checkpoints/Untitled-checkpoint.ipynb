# 决策树 Decision Tree

决策树是根据某个属性判断是否，以此来形成一棵树。决策树是一种`绝对分类`，从训练集训练得到的决策树会在训练集中拥有极好的训练效果。

决策树不仅可以用于分类，也可以用于回归：
* 分类是根据属性节点的信息熵，选择信息熵较小的属性作为节点，最后根据叶节点中占据多数的类别作为最终结果。
* 回归是根据属性节点的方差，选择划分后方差较小的属性作为节点，最后取叶节点的平均值作为回归值。

## 信息熵

普通的决策树判断依据是信息熵，信息熵是用来衡量一个集合中信息的混乱程度，也就是该节点的信息能够表达类别的准确程度。

$$
Ent(D) = -\sum^n_{k=1} p_klog_2p_k
$$

值越大数据越混乱，也就越难作为节点进行决策。


## 剪枝（过拟合）

对决策树进行分类的时候，往往会由于树的深度太深，从而出现过拟合的情况，为了解决该问题，需要对决策树进行剪枝操作，一般来说有预剪枝和后剪枝两种操作：
* 预剪枝：在创建决策树的同时进行剪枝。设置一个阈值对决策树节点的信息熵进行判断，若该点的信息熵小于设立的阈值（不能带来有效划分）则停止创建分枝。
* 后剪枝：在创建完决策树之后对决策树进行剪枝，从叶节点开始依次对其父结点进行检查，根据合并后的信息熵增加量是否小于阈值，来决定是否将叶节点进行合并。

## 决策树的改进

### ID3决策树

ID3决策树是最一般的决策树，基于信息熵进行判断，选择`信息增益量最大`的属性作为决策节点。

$$
    Gain(A) = Info(D) - Info_A(D)
$$

Info(D)是未知属性时节点的信息熵，Info_A(D)是使用A属性作为分类标准后节点的信息熵。

### C4.5决策树（在其中使用较多）

基于ID3决策树进行改进，选择`信息增益率最大`的属性作为决策节点。

$$
    GainRatio(A) = \frac{Gain(A)}{SplitInfo(A)}
$$

$$
    SplitInfo(A) = - \sum^n_{j=1}\frac{\lvert D_j \rvert}{\lvert D \rvert} \times log_2(\frac{\lvert D_j \rvert}{\lvert D \rvert})
$$

[可以参考此处的C4.5决策树举例说明](https://blog.csdn.net/zjsghww/article/details/51638126?utm_medium=distribute.pc_relevant.none-task-blog-BlogCommendFromBaidu-1.channel_param&depth_1-utm_source=distribute.pc_relevant.none-task-blog-BlogCommendFromBaidu-1.channel_param)

### CART决策树

分类回归树通过利用GINI指数（与信息熵概念相似），选择`GINI_Gain越小`的属性作为决策节点。

$$
GINI\_Gain(A)= \sum^n_{j=1} \frac{\lvert D_j \rvert}{\lvert D \rvert} \times (1-\sum(\frac{\lvert C_i \rvert}{\lvert D_j \rvert})^2)
$$

D是各个属性的总数，D_j是该属性中某个取值的样本数，C_i是D_j中属于某类别的样本数

```python
%matplotlib inline
# %matplotlib widget
plt.rcParams["font.sans-serif"] = 'SimHei' # 中文问题
plt.rcParams["axes.unicode_minus"]= False # 负号问题
%config InlineBackend.figure_format = 'svg'
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Parameters
n_classes = 3
plot_colors = "ryb"
plot_step = 0.02

# Load data
iris = load_iris()

for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3], [1, 2], [1, 3], [2, 3]]):
    # We only take the two corresponding features 2维便于直观感受
    X = iris.data[:, pair]
    y = iris.target

    # Train
    clf = DecisionTreeClassifier().fit(X, y)

    # Plot the decision boundary
    plt.subplot(2, 3, pairidx + 1)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu)

    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])

    # Plot the training points
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color, label=iris.target_names[i],
                    cmap=plt.cm.RdYlBu, edgecolor='black', s=15)

plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend(loc='lower right', borderpad=0, handletextpad=0)
plt.axis("tight")

plt.figure()
clf = DecisionTreeClassifier().fit(iris.data, iris.target)
plot_tree(clf, filled=True)
plt.show()
```
