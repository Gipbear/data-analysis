## 索引操作

```python
# 从字典对象创建DataFrame，并设置索引
dates = pd.date_range('today', periods=10)
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no'],
        'dates': dates,
        'alpha': list('qwertyuiop'),
        'index1': list(np.random.permutation(10))
        }
indexs = [list('abcdefghij')]

df = pd.DataFrame(data=data, index=indexs)
df
```

```python
# 输出所有列名
df.columns
```

```python
# 动物的种类
df['animal'].unique()
```

```python
# 修改dates为datetimes
df.rename(columns={'dates': 'datetimes'}, inplace=True)
```

```python
# 显示df的基础信息，包括行的数量；列名；每一列值的数量、类型
df.info()
# df.describe()
```

```python
# 第1，5，7行的 animal 和 age 列
df.loc[df.index[[0, 4, 6]], ['animal', 'age']]
```

```python
# 年龄为2，3，5的行
df[df['age'].isin([2, 3, 5])]
```

```python
# 插入新行 k 为'dog', 5.5, 2, 'no',当前时间, l, 100，并删除
df.loc['k', :] = ['dog', 5.5, 2, 'no', datetime.now(), 'l', 100]
df = df.drop('k')
df
```

```python
# 取出 age 缺失的行
df[df['age'].isnull()]
```

```python
# 计算 visit 列总和
df['visits'].sum()
```

```python
# 计算 animal 各个类别出现的次数
df['animal'].value_counts()
```

```python
# 计算各类 animal 的平均年龄
df.groupby('animal')['age'].mean()
```

```python
# 按照 visits 列升序，再按照 age 列降序
df.sort_values(by=['visits', 'age'], ascending=[True, False])  # inplace=True
```

```python
# 将 priority 中的值替换为布尔值
df['priority'] = df['priority'].map({'yes': True, 'no': False})
df['priority']
```

```python
# 将其中的 cat 替换为 python
df = df.replace('snake', 'camel')
```

```python
# 利用上下值填充NAN
df['age'] = df['age'].fillna(df['age'].interpolate())
df
```

```python
# 计算每种 animal 的不同 visit 数量的平均 age
df.pivot_table(index='animal', columns='visits', values='age', aggfunc='mean')
```

```python
# 删除 animal 重复值
df.drop_duplicates(subset='animal')
```

```python
# visits 最大值所在的行
df[df['visits'] == df['visits'].max()]
```

```python
# 统计 animal 每个字符串的长度
df['animal'].map(lambda x: len(x))
```

```python
# 将日期转化为年-月-日 小时-分钟-秒
for i in range(df.shape[0]):
    df.iloc[i, 4] = df.iloc[i, 4].strftime('%Y-%m-%d %H:%M:%S')
df
```

```python
df.dtypes
```

```python
# 删除最后 alpha 列
del df['alpha']
```

```python
# 合并1、2列
df['1h2'] = df['animal'] + '_' + df['age'].map(str)  # int需要先转化为str
df
```

```python
# 再将其进行拆分
df['new'] = df['1h2'].str.split('_')
df['new']
```

## 数值操作

```python
df = pd.DataFrame(np.random.randn(
    4, 5), index=range(2, 6), columns=list('abcde'))
df
```

```python
# df2 是 df1 的各个数字减去该行均值
df.sub(df.mean(axis=1), axis=0)
```

```python
# 找出列之和最小的一列
print(df.sum())
df.sum().idxmin()
```

```python
# 给定DataFrame，求A列每个值的前3个对应B列的值的和
df = pd.DataFrame({'A': list('aaabbcaabcccbbc'),
                   'B': [12, 345, 3, 1, 45, 14, 4, 52, 54, 23, 235, 21, 57, 3, 87]})
df.groupby('A')['B'].nlargest(3).sum(level='A')
```

```python
# 对A列每10步进行分箱，求对应B列的和
df = pd.DataFrame({'A': [1, 2, 11, 11, 33, 34, 35, 40, 79, 99],
                   'B': [1, 2, 11, 11, 33, 34, 35, 40, 79, 99]})
df.groupby(pd.cut(df['A'], np.arange(0, 101, 10)))['B'].sum()
```

```python
# 有点🚹，计算每个元素距离左侧最近的0的距离(第一个默认左侧为0)，生成新列Y
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
x = (df['X'] != 0).cumsum()
y = x != x.shift()
(y != y.shift()).cumsum()  # 以0为间断点进行分组
df['Y'] = y.groupby((y != y.shift()).cumsum()).cumsum()
df
```

```python
# 找出其中最大的三个值的坐标
df = pd.DataFrame(np.random.random(size=(5, 3)), columns=['a', 'b', 'c'])
print(df)
df.T.unstack().sort_values()[-3:].index.tolist()
```

```python
# 将小数转化为百分号
df[['a']].style.format({'a': '{0:.2%}'.format})
```

```python
# X 下一行比上一行的差值
df = pd.DataFrame({'X': [7, 2, 10, 3, 4, 2, 5, 0, 3, 4]})
df['X'].diff()
```

```python
# X 下一行比上一行的变化率
df['X'].pct_change()
```

```python
# 查看数据的最小值，25%，50%，75%，以及最大值
np.percentile(df, q=[0, 25, 50, 75, 100])
```

```python
# 将其中的负数替换为该组均值
df = pd.DataFrame({'grps': list('aaabbcaabcccbbc'),
                   'vals': [-12, 345, 3, 1, 45, 14, 4, -52, 54, 23, -235, 21, 57, 3, 87]})


def replace(group):
    mask = group < 0
    group[mask] = group[~mask].mean()
    return group


df['vals'] = df.groupby('grps')['vals'].transform(replace)
df
```

```python
# 计算3位滑动窗口的平均值，忽略NAN
df = pd.DataFrame({'group': list('aabbabbbabab'),
                   'value': [1, 2, 3, np.nan, 2, 3, np.nan, 1, 7, 3, np.nan, 8]})
g1 = df.groupby('group')['value']
g2 = df.fillna(0).groupby('group')['value']
s = g2.rolling(3, min_periods=1).sum() / g1.rolling(3, min_periods=1).count()
s.reset_index(level=0, drop=True).sort_index()
```

```python
# 第一列中不在第二列的数字
df = pd.DataFrame({'A': [3, 3, 5, 7, 1, 5, 4], 'B': [
                  2, 5, 6, 7, 7, 4, 7]}, index=list('abcdefg'))
df['A'][~df['A'].isin(df['B'])]
```

```python
# 第一列和第二列中出现频率最高的前三个数字
temp = df['A'].append(df['B'])
temp.value_counts().index[:3]
```

```python
# 查找第A列局部最大值的位置，即该数字比前面和后面都大
tem = np.diff(np.sign(np.diff(df['A'])))
np.where(tem == -2)[0] + 1
```

```python
# 计算两列之间的欧式距离
np.linalg.norm(df['A']-df['B'])  # 计算norm默认参数就是计算矩阵内所有元素的平方和开根号
```

```python
# 对多列数据进行合并
df1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                    'key2': ['K0', 'K1', 'K0', 'K1'],
                    'A': [1, 2, 4, 7],
                    'B': [3, 5, 2, 7]})

df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                    'key2': ['K0', 'K0', 'K0', 'K0'],
                    'C': [4, 7, 2, 9],
                    'D': [8, 3, 1, 6]})

pd.merge(df1, df2, on=['key1', 'key2'])
```
